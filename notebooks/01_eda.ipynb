```python
import sys
sys.path.append('..')  # to import src modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.data.loader import load_raw_data, save_processed_data
from src.utils.logger import setup_logger
logger = setup_logger('eda')
%matplotlib inline

df = load_raw_data('../data/raw/online_retail.xlsx')

# Remove rows without CustomerID
df_clean = df.dropna(subset=['CustomerID']).copy()
# Remove rows with negative Quantity or UnitPrice (or treat as returns)
df_clean = df_clean[(df_clean['Quantity'] > 0) & (df_clean['UnitPrice'] > 0)]
# Convert CustomerID to int
df_clean['CustomerID'] = df_clean['CustomerID'].astype(int)
# Create TotalPrice
df_clean['TotalPrice'] = df_clean['Quantity'] * df_clean['UnitPrice']

save_processed_data(df_clean, '../data/processed/cleaned_data.csv')
```
